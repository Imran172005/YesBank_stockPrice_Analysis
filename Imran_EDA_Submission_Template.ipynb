## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Parse Date column
def parse_date(d):
    month_map = {
        'Jan':'01','Feb':'02','Mar':'03','Apr':'04','May':'05','Jun':'06',
        'Jul':'07','Aug':'08','Sep':'09','Oct':'10','Nov':'11','Dec':'12'
    }
    parts = d.split('-')
    return f'{parts[1]}-{month_map[parts[0]]}-01'

df['Parsed_Date'] = df['Date'].apply(parse_date)
df['Parsed_Date'] = pd.to_datetime(df['Parsed_Date'])
df['Year'] = df['Parsed_Date'].dt.year
df['Month'] = df['Parsed_Date'].dt.month_name()
df['Month_Num'] = df['Parsed_Date'].dt.month

# Derived features
df['Price_Range'] = df['High'] - df['Low']
df['Price_Change'] = df['Close'] - df['Open']
df['Daily_Return'] = df['Close'].pct_change() * 100
df['Close_MA_12'] = df['Close'].rolling(window=12).mean()

# Sort by date
df = df.sort_values('Parsed_Date').reset_index(drop=True)

print('Data wrangling complete. Shape:', df.shape)
df[['Date','Open','High','Low','Close','Year','Month','Price_Range','Daily_Return']].head()
````

### What all manipulations have you done and insights you found?

**Manipulations performed:**
1. Parsed the Date column from 'Mon-YY' format to proper datetime
2. Extracted Year, Month, and Month_Num features
3. Created Price_Range (High - Low) to measure monthly volatility
4. Created Price_Change (Close - Open) to see monthly direction
5. Calculated Daily_Return as percentage change in Close price
6. Added 12-month rolling average (Close_MA_12) for trend analysis
7. Sorted chronologically

**Key Insights:**
- The stock had massive growth from ~13 (2005) to ~404 (Aug 2018)
- A dramatic collapse followed, dropping to ~5.55 (Mar 2020 low)
- The stock is highly volatile with wide price ranges in certain monthsAnswer Here.

#### Chart - 14 - Correlation Heatmap

````python
# Chart - 15 visualization code
plt.figure(figsize=(10, 8))
numeric_cols = ['Open', 'High', 'Low', 'Close', 'Price_Range', 'Price_Change', 'Daily_Return']
corr = df[numeric_cols].corr()
mask = np.triu(np.ones_like(corr, dtype=bool), k=1)
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='RdBu_r', center=0,
            square=True, linewidths=0.5, cbar_kws={'shrink': 0.8})
plt.title('Correlation Heatmap of Numeric Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
````

**1. Why did you pick the specific chart?**
Heatmap visualizes the correlation matrix, showing relationships between all numeric variables (Multivariate).

**2. What is/are the insight(s) found from the chart?**
Open, High, Low, Close are all very highly correlated (>0.99) as expected. Price_Range has moderate correlation with price levels. Daily_Return and Price_Change show low correlation with price levels, indicating returns are relatively independent of price magnitude.
